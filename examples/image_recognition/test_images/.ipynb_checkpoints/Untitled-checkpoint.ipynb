```python
import cv2
import PIL
import numpy as np
```

```python
img = PIL.Image.open('basic-color-circles-printable-1.jpg')
```

```python
img
```

```python
arr = np.array(img)
```

```python
arr
```

```python
hsv = cv2.cvtColor(arr, cv2.COLOR_RGB2HSV)
hsv.shape
```

```python
maxx = np.amin(np.amin(hsv, axis=0), axis=0)
maxx.shape
```

```python
maxx
```

```python
boundaries = [[(np.array([0, 20, 5]), np.array([10, 255, 255]))], [(np.array([170, 20, 5]), np.array([180, 255, 255]))]]
```

```python
for (lower, upper) in boundaries[0]:
    mask1 = cv2.inRange(hsv, lower, upper)
for (lower, upper) in boundaries[1]:
    mask2 = cv2.inRange(hsv, lower, upper)
mask = mask1 | mask2
```

```python
PIL.Image.fromarray(mask)
```
