```python
import cv2
import PIL
import numpy as np
```

```python
img = PIL.Image.open('ban_app.jpg')
```

```python
img
```

```python
arr = np.array(img)
```

```python
arr
```

```python
hsv = cv2.cvtColor(arr, cv2.COLOR_RGB2HSV)
hsv.shape
```

```python
maxx = np.amin(np.amin(hsv, axis=0), axis=0)
maxx.shape
```

```python
maxx
```

```python
boundaries = [[(np.array([20, 20, 5]), np.array([30, 255, 255]))], [(np.array([40, 10, 5]), np.array([70, 255, 255]))]]
```

```python
for (lower, upper) in boundaries[0]:
    mask1 = cv2.inRange(hsv, lower, upper)
for (lower, upper) in boundaries[1]:
    mask2 = cv2.inRange(hsv, lower, upper)
mask = mask1 | mask2
```

```python
PIL.Image.fromarray(mask)
```
